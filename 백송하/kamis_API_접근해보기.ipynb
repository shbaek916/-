# 사과로 월별 가격 데이터 요청하고 정리해보기

```python
url = 'http://www.kamis.or.kr/service/price/xml.do?action=monthlySalesList'

param = {
    'p_cert_key' : '8dc40ba8-e389-4163-98df-2ea1390c88f6',
    'p_cert_id' : '2455',
    'p_returntype' : 'json',
    'p_yyyy' : '2022',
    'p_period' : '5', # 17~22년 까지 5개년치
    'p_itemcategorycode': '400', # 과일류
    'p_kindcode' : '05', # 후지
    'p_graderank' : '1', # 상급
    'p_countycode' : '1101',  # 디폴트 전체지역
    'p_convert_kg_yn': 'Y',
    'p_itemcode' : '411' # 사과
}
header = {
    'p_cert_key' : '8dc40ba8-e389-4163-98df-2ea1390c88f6',
    'p_cert_id' : '2455'
}
resp = requests.get(url = url, params = param, headers = header)
resp.text
```

요청한 결과 json 파일로 따로 저장 후 딕셔너리 타입의 변수로 저장

```python
with open('apple.json', 'w', encoding='utf8') as f:
    f.write(resp.text)
with open('apple.json', 'r') as f:
  apple_json = json.load(f)
apple_json
```

```python
apple_json['price']
```

```python
data = []
for j in range(5):
  data.append([])
  data[j].append(apple_json['price'][0]['item'][j]['yyyy'])
  for i in range(1,13):
    try : 
      data[j].append(int(apple_json['price'][0]['item'][0][f'm{i}'].replace(',', '')))
    except: 
      data[j].append(0)

apple_huji = pd.DataFrame(data)
col = [ f'{i}월' for i in range(1,13)]
col.insert(0,'연도')
apple_huji.columns = col
apple_huji
```
